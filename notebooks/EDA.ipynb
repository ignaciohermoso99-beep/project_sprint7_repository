```python
import pandas as pd
import numpy as np
import plotly.express as px
pd.set_option('display.max_columns', 100)
```

```python
car_data = pd.read_csv('../vehicles_us.csv',parse_dates=['date_posted'], low_memory=False)
```

```python
car_data.info()
```

```python
car_data.describe()
```

```python
car_data.head(10)
```

```python
print('Los valores ausentes dentro del notebook car_data son:')
car_data.isna().sum().sort_values(ascending = False).head()
```

```python
car_data['price'] = car_data['price'].astype(float)
car_data['model_year'] = car_data['model_year'].fillna(0).astype(int)
```

```python
car_data.info()
```

```python

car_data['condition'] = car_data['condition'].str.lower().str.strip()
car_data['model'] = car_data['model'].str.lower().str.strip()
```

```python
car_data.duplicated().sum()
```

```python
car_data['age'] = 2025 - car_data['model_year']
car_data['year_posted'] = car_data['date_posted'].dt.year
```

```python
car_data.head()
```

```python
fig = px.histogram(car_data, x="odometer")
fig.show() # crear grfico de dispersin
```

```python
fig = px.scatter(car_data, x="odometer", y="price") # crear un grfico de dispersin
fig.show() # crear grfico de dispersin
```
